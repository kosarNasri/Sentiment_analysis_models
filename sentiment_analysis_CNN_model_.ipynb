### Preprocessing


```python
import re
import hazm  # Install using: pip install hazm
from hazm import Normalizer, word_tokenize, stopwords_list

def preprocess_text(text):
    # Normalize text
    normalizer = Normalizer()
    normalized_text = normalizer.normalize(text)

    # Tokenize text
    tokens = word_tokenize(normalized_text)

    # Remove stopwords
    # tokens = [token for token in tokens if token not in stopwords_list()]

    # Remove punctuation and non-alphabetic characters
    tokens = [re.sub(r'[^a-zA-Zآ-ی۰-۹]', '', token) for token in tokens]

    # Remove empty tokens
    tokens = [token for token in tokens if token != '']

    # Join tokens back into a single string
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text
```

```python
import torch
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator

# Define a tokenizer function
tokenizer = get_tokenizer('basic_english')

# Sample comments
comments = ['نظر شما چیست؟', 'این یک نمونه است', 'تست توکنایزر']

# Tokenize comments
tokenized_comments = [tokenizer(comment) for comment in comments]

# Build vocabulary from the tokenized comments
vocab = build_vocab_from_iterator(tokenized_comments, specials=["<unk>"])
vocab.set_default_index(vocab["<unk>"])

# Convert comments to tensor of numericalized tokens
numericalized_comments = [torch.tensor(vocab(tokenized_comment)) for tokenized_comment in tokenized_comments]

print(numericalized_comments)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['comment'],data['recommendation_status'], test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_train.shape)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
max_words=10000
tokenizer=Tokenizer(max_words)
tokenizer.fit_on_texts(X_train)
sequence_train=tokenizer.texts_to_sequences(X_train)
sequence_test=tokenizer.texts_to_sequences(X_test)
```

```python
word2vec=tokenizer.word_index
V=len(word2vec)
print('dataset has %s number of independent tokens' %V)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
data_train=pad_sequences(sequence_train)
data_train.shape
```

```python

T=data_train.shape[1]
data_test=pad_sequences(sequence_test,maxlen=T)
data_test.shape
```

```python
y_train = y_train.values
y_test = y_test.values

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.layers import Input,Conv1D,MaxPooling1D,Dense,GlobalMaxPooling1D,Embedding
from tensorflow.keras.models import Model

# Importing libraries
from keras.models import Sequential
from keras.layers import Flatten
from keras.preprocessing import sequence
```

```python
D=20
i=Input((T,))
x=Embedding(V+1,D)(i)
x=Conv1D(32,2,activation='relu')(x)
x=MaxPooling1D(3)(x)
x=Conv1D(64,2,activation='relu')(x)
x=MaxPooling1D(2)(x)
x=Conv1D(128,2,activation='relu')(x)
x=GlobalMaxPooling1D()(x)
x=Dense(2,activation='softmax')(x)
model=Model(i,x)
model.summary()
```

```python
#EarlyStopping and ModelCheckpoint
from keras.callbacks import EarlyStopping, ModelCheckpoint
es = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 5)
mc = ModelCheckpoint('./model.h5', monitor = 'val_accuracy', mode = 'max', verbose = 1, save_best_only = True)
```

```python
data_train[0]
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
cnn_senti=model.fit(data_train,y_train,validation_data=(data_test,y_test),epochs=30,batch_size=128,verbose = 1,callbacks = [es,mc])
```

```python
y_pred=model.predict(data_test)
y_pred
```

```python
import numpy as np
y_pred=np.argmax(y_pred,axis=1)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
cm=confusion_matrix(y_test,y_pred)
ax=sns.heatmap(cm,annot=True,cmap='Blues',fmt=' ')
ax.set_title('Confusion Matrix')
ax.set_xlabel('y_test')
ax.set_ylabel('y_pred')
```
